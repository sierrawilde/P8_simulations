```python
def LoadFilelist(mypath, job_id, search_str = '.'):
    '''
    Get list of all files fullfilling a few naming conditions
    '''
    filelist = []
    #print('Searching files in {} from run {} with "{}" in filename'.format(mypath, job_id, search_str))
    for (dirpath, dirnames, filenames) in os.walk(mypath, topdown=True):
        for name in filenames:
            if str(job_id) in dirpath and search_str in name and 'zero' not in name:
                filelist.append(os.path.join(dirpath, name))
    return filelist

def collect_and_sort_job_output_file(dict_path_list, job_range):
    mega_dict = {}

    for dict_path in dict_path_list:
        path_splitted = dict_path.split('/')
        job_id = int(path_splitted[-2])

        if job_id >= job_range[0] and job_id<= job_range[1]:


            with open(dict_path, 'r') as infile:
                job_output = json.load(infile)            

            for k in job_output.keys():
                global_id = int(str(job_id)+str(k))
                #print(global_id)
                mega_dict[global_id] = job_output[k]
                
                #print(len(job_output[k]['start_times_true']))

    return mega_dict
```

```python
job_output_path  ='../20200626_sierra/'
job_range = [926379, 926878]

mega_dict_path_list = LoadFilelist(job_output_path, '', 'mega_dict' )
mega_dict = collect_and_sort_job_output_file(mega_dict_path_list, job_range)
```

```python
with open('../20200626_sierra/mega_dict.json', 'w') as outfile:
    json.dump(mega_dict, outfile)
```
